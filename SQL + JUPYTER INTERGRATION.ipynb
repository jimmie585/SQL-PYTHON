```python
import pyodbc
import pandas as pd

# Use double backslashes (\\) or raw string (r'')
server = r'ANALYSTJIMMIE\SQLEXPRESS'  # Example: 'LAPTOP\\SQLEXPRESS' or 'localhost'
database = 'portfolio project'
username = ''  # Leave blank for Windows Authentication
password = ''  # Leave blank for Windows Authentication
driver = '{ODBC Driver 17 for SQL Server}'

try:
    # Use Trusted Connection for Windows Authentication
    conn = pyodbc.connect(f'DRIVER={driver};SERVER={server};DATABASE={database};Trusted_Connection=yes;')
    print("Connected to the database successfully!")
except Exception as e:
    print("Error:", e)
```

```python
# Query execution
query = "SELECT * FROM Employees;"  
df = pd.read_sql(query, conn)

# Display the first few rows
df.head()
```

```python
cursor = conn.cursor()

insert_query = """
INSERT INTO Employees (employee_id, first_name) 
VALUES (?, ?)"""  

values = ('3', 'jane')  
cursor.execute(insert_query, values)

conn.commit()  # Save changes
print("Data inserted successfully!")
```
